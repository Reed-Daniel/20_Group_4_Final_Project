#Cleaning of Players DataFrame complete.

```python
# Find % missing data for player_data DF
for col in player_data.columns:
    pct_missing = np.mean(player_data[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing * 100)))
```

```python
empty_player_data = player_data.isnull().sum()
empty_player_data
```

```python
# Filter player_data to players who started after the 1979 season when the 3 point line was added.
player_data = player_data.loc[player_data['year_start'] >=1980]

#modern_season_stats = season_stats.loc[season_stats['Year'] >=1980]
```

```python
# Find rows in player_data DF that are empty
null_player_data = player_data[player_data.isna().any(axis=1)]
null_player_data
# Empty values are already NaN so will leave as is
```

```python
# Show player_data DF data types
player_data.dtypes
```

```python
# Turn height column into height(inches) using Lambda function

# 1) Change data type to string to use .split()
player_data['height(inches)'] = player_data['height'].astype('string')
# 2) Use lambda function to turn height into inches applying separate formulas for height and inch conversion
player_data['height(inches)'] = player_data['height(inches)'].apply(lambda x: int(x.split('-')[0])*12 + int(x.split('-')[1]*1))
```

```python
player_data.head()
```

```python
# Take position column and split into just their primary position (the first one)

# 1) Change data type to string
player_data['position'] = player_data['position'].astype('string')
# 2) Use lambda to split, leaving only their primary position (first listed)
player_data['position'] = player_data['position'].str.split('-').str[0]
```

```python
# Reorder columns in player_data
player_data = player_data.reindex(['name', 'year_start', 'year_end', 'position', 'height', 'height(inches)', 'weight', 'birth_date', 'college'], axis=1)
```

```python
# Convert player_data.weight to integer
player_data['weight'] = player_data['weight'].astype(int)
```

```python
player_data.head()
```

```python
player_data.dtypes
```

```python
# Convert objects to correct values for Postgres
player_data['name'] = player_data['name'].astype('string')
player_data['position'] = player_data['position'].astype('string')
player_data['height'] = player_data['height'].astype('string')
player_data['birth_date'] = player_data['birth_date'].astype('string')
player_data['college'] = player_data['college'].astype('string')
```

```python
# Remove special characters from name column
player_data['name'] = player_data['name'].str.replace('*','')
```

#Data cleaning for player_data DF complete

```python
# Find % missing data for season_stats DF
for col in season_stats.columns:
    pct_missing = np.mean(season_stats[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# List all columns for season_stats dataframe
season_stats.columns
```

```python
#For season_stats DF, drop players Unnamed: 0, blanl, blank2, columns
season_stats = season_stats.drop(columns=['Unnamed: 0', 'blanl', 'blank2'])
```

```python
# Make modern_season_stats DF for seasons after 3-point line was added (1980 onwards)
modern_season_stats = season_stats.loc[season_stats['Year'] >=1980]
```

```python
# Find % missing data for modern_season_stats DF
for col in modern_season_stats.columns:
    pct_missing = np.mean(modern_season_stats[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100000)))
```

```python
# Replace empty stats with 0s.
modern_season_stats = modern_season_stats.fillna(0)
```

```python
# Check to see NaNs replaced with 0
for col in modern_season_stats.columns:
    pct_missing = np.mean(modern_season_stats[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100000)))
```

```python
# Add row_id to modern_season_stats to identify unique rows for SQL primary key
modern_season_stats['id'] = np.arange(len(modern_season_stats))
```

```python
# Remove special characters from Player column
modern_season_stats['Player'] = modern_season_stats['Player'].str.replace('*','')
```

```python
modern_season_stats.head(2)
```

```python
modern_season_stats.dtypes
```

```python
# Fix modern_season_stats objects for Postgres
modern_season_stats['Year'] = modern_season_stats['Year'].astype(int)
modern_season_stats['Player'] = modern_season_stats['Player'].astype('string')
modern_season_stats['Pos'] = modern_season_stats['Pos'].astype('string')
modern_season_stats['Age'] = modern_season_stats['Age'].astype(int)
modern_season_stats['Tm'] = modern_season_stats['Tm'].astype('string')
modern_season_stats['G'] = modern_season_stats['G'].astype(int)
modern_season_stats['GS'] = modern_season_stats['GS'].astype(int)
modern_season_stats['MP'] = modern_season_stats['MP'].astype(int)
modern_season_stats['FG'] = modern_season_stats['FG'].astype(int)
modern_season_stats['FGA'] = modern_season_stats['FGA'].astype(int)
modern_season_stats['3P'] = modern_season_stats['3P'].astype(int)
modern_season_stats['3PA'] = modern_season_stats['3PA'].astype(int)
modern_season_stats['2P'] = modern_season_stats['2P'].astype(int)
modern_season_stats['2PA'] = modern_season_stats['2PA'].astype(int)
modern_season_stats['FT'] = modern_season_stats['FT'].astype(int)
modern_season_stats['FTA'] = modern_season_stats['FTA'].astype(int)
modern_season_stats['ORB'] = modern_season_stats['ORB'].astype(int)
modern_season_stats['DRB'] = modern_season_stats['DRB'].astype(int)
modern_season_stats['TRB'] = modern_season_stats['TRB'].astype(int)
modern_season_stats['AST'] = modern_season_stats['AST'].astype(int)
modern_season_stats['STL'] = modern_season_stats['STL'].astype(int)
modern_season_stats['BLK'] = modern_season_stats['BLK'].astype(int)
modern_season_stats['TOV'] = modern_season_stats['TOV'].astype(int)
modern_season_stats['PTS'] = modern_season_stats['PTS'].astype(int)
```

#Finished cleaning modern_season_stats DF

```python
# Make DF of statistics for each game
per_game_stats = modern_season_stats[['Year', 'Player', 'Pos']]
#per_game_stats = modern_season_stats['Player']
#per_game_stats = modern_season_stats['Pos']
```

```python
# Add per-game statistics dataframe (stat / games)

# Minutes per game
per_game_stats['MP(Game)'] = (modern_season_stats['MP'] / modern_season_stats['G']).round(1)
# Points per game
per_game_stats['Points(Game)'] = (modern_season_stats['PTS'] / modern_season_stats['G']).round(1)
# Assists per game
per_game_stats['Assists(Game)'] = (modern_season_stats['AST'] / modern_season_stats['G']).round(1)
# Rebounds per game
per_game_stats['Rebounds(Game)'] = (modern_season_stats['TRB'] / modern_season_stats['G']).round(1)
# Steals per game
per_game_stats['Steals(Game)'] = (modern_season_stats['STL'] / modern_season_stats['G']).round(1)
# Blocks per game
per_game_stats['Blocks(Game)'] = (modern_season_stats['BLK'] / modern_season_stats['G']).round(1)
# Turnovers per game
per_game_stats['Turnover(Game)'] = (modern_season_stats['TOV'] / modern_season_stats['G']).round(1)
```

```python
# Add row_id to per_game_stats to identify unique rows for SQL primary key
per_game_stats['id'] = np.arange(len(per_game_stats))
```
