Data Cleaning

```python
# Correcting column name in players DF
players.rename(columns={'collage':'college'}, inplace=True)
```

```python
#players.head()
```

```python
# Change players 'weight' column from kilograms to pounds.
players['weight'] = players['weight'] / 2.2046
players.round(0)
```

```python
# Find % missing data for players DF
for col in players.columns:
    pct_missing = np.mean(players[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Find % missing data for player_data DF
for col in player_data.columns:
    pct_missing = np.mean(player_data[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```
